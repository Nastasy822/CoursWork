создание БД

```
#номмализация + предобработка
import numpy as np
import face_recognition as fr 
import json
import glob

sp = dlib.shape_predictor('/content/drive/MyDrive/myrsach/face_detection/shape_predictor_68_face_landmarks.dat')
facerec = dlib.face_recognition_model_v1('/content/drive/MyDrive/myrsach/face_detection/dlib_face_recognition_resnet_model_v1.dat')
detector = dlib.get_frontal_face_detector()

for file in glob.glob("/content/drive/MyDrive/myrsach/face_detection/images/*"):
    
  
    identity = os.path.splitext(os.path.basename(file))[0]
    
    # proccessing the second picture
    print(file)
    image = cv2.imread(file)
    faces = detector(image, 1)
    shape = sp(image, faces[0])
    face_descriptor = facerec.compute_face_descriptor(image, shape)

    ebm  = np.asarray(face_descriptor)

    ebm =  ebm/np.sqrt(np.sum(ebm**2))
    ebm=ebm.tolist()
    to_json = {'name': identity, 'embeding': ebm}
    with open('students.json', 'a') as f:
        f.write(json.dumps(to_json))
        f.write('\n')
```
